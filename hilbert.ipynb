```python
import functools
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
import itertools
from matplotlib.colors import ListedColormap
```

```python
hilbert_map = {
    'a': {(0, 0): (0, 'd'), (0, 1): (1, 'a'), (1, 0): (3, 'b'), (1, 1): (2, 'a')},
    'b': {(0, 0): (2, 'b'), (0, 1): (1, 'b'), (1, 0): (3, 'a'), (1, 1): (0, 'c')},
    'c': {(0, 0): (2, 'c'), (0, 1): (3, 'd'), (1, 0): (1, 'c'), (1, 1): (0, 'b')},
    'd': {(0, 0): (0, 'a'), (0, 1): (3, 'c'), (1, 0): (1, 'd'), (1, 1): (2, 'd')},
}
def point_to_hilbert(order, x, y):
    current_square = 'a'
    position = 0
    for i in range(order - 1, -1, -1):
        position <<= 2
        quad_x = 1 if x & (1 << i) else 0
        quad_y = 1 if y & (1 << i) else 0
        quad_position, current_square = hilbert_map[current_square][(quad_x, quad_y)]
        position |= quad_position
    return position
hilbert_32 = functools.partial(point_to_hilbert,5)
```

```python
def twod_grid(n, size=1):
    """given a size, generate a list of pairs of coordinates
    corresponding with grid points on a sizexsize grid"""
    xs = ys = np.arange(start=0,stop=int(n*size), step=size)
    return list(itertools.product(xs, ys))
```

```python
grid = twod_grid(32)
with_score = [(x,y, hilbert_32(x,y)) for x,y in grid]
pal = sns.color_palette("RdPu")
my_cmap = ListedColormap(colors=pal)
fig, ax = plt.subplots(1)
ax.scatter([i[0] for i in with_score], [i[1] for i in with_score],
           c=[i[2] for i in with_score],cmap=my_cmap)
```

```python
in_hilber_order = sorted(grid, key= lambda i:hilbert_32(*i))
fig, ax = plt.subplots(1)
ax.plot([i[0] for i in in_hilber_order], [i[1] for i in in_hilber_order], label='1D Hilbert curve')
ax.set_xlabel('Node x coordinate')
ax.set_ylabel('Node y coordinate')
sns.despine()
fig.savefig('Hilbert_Curve', format='png')
#ax.legend(loc='best')
```
